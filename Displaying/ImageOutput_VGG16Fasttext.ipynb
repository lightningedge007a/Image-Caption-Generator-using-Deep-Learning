```
def image_caption(idx,dic):
    fname=dataset+idx+'.jpg'
    img1=cv2.imread(fname)
    img2=cv2.cvtColor(img1,cv2.COLOR_RGB2BGR)
    plt.pause(0.1)
    fig=plt.figure()
    fig,axarr = plt.subplots(1,2)
    axarr[0].imshow(img1)
    axarr[1].imshow(img2)
    #plt.imshow(img)
    print("references :")
    for cap in dic[idx][0]:
        print_caption(cap) 
    print("caption generated :")
    print_caption(dic[idx][1])
```

```
def print_caption(l):
        s=''
        for x in l[1:-1]:
            s=s+" "+str(x)
        print(s+"\n")
```

```
for i in l1:
    image_caption(i,rc1)
for i in l2:
    image_caption(i,rc2)
for i in l3:
    image_caption(i,rc3)
```
